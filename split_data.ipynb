```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python

INPUT_TRAIN_CSV = 'DanishFungi2024-Mini-train.csv'  
INPUT_TEST_CSV  = 'DanishFungi2024-Mini-pubtest.csv'       
OUTPUT_TRAIN    = 'final_train_80.csv'
OUTPUT_VAL      = 'final_val_10.csv'
RANDOM_SEED = 42

def split_dataset_by_count():
    # Load test set
    test_df = pd.read_csv(INPUT_TEST_CSV)
    val_count = len(test_df)
    print(f"Target Validation Size (from PubTest): {val_count} images")

    # Load the Training data
    train_source_df = pd.read_csv(INPUT_TRAIN_CSV)
    total_train_source = len(train_source_df)
    print(f"Training Data: {total_train_source} images")

    # Perform Split
    print(f"Splitting {val_count} images out for Validation...")
    
    train_df, val_df = train_test_split(
        train_source_df,
        test_size=val_count, 
        stratify=train_source_df['class_id'], 
        random_state=RANDOM_SEED
    )

    # Save Files
    train_df.to_csv(OUTPUT_TRAIN, index=False)
    val_df.to_csv(OUTPUT_VAL, index=False)

    # Calculate Ratios
    total_imgs = len(train_df) + len(val_df) + len(test_df)
    r_train = len(train_df) / total_imgs * 100
    r_val = len(val_df) / total_imgs * 100
    r_test = len(test_df) / total_imgs * 100
    
    print(f"Data Ratios: {r_train:.1f}% / {r_val:.1f}% / {r_test:.1f}%")

if __name__ == "__main__":
    split_dataset_by_count()
```
